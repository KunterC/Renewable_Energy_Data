```python
import pandas as pd #imports pandas, pyplot from matplotlib and seaborn
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
energy_df = pd.read_csv("Renewable_Energy_Adoption.csv") #reads the data
energy_df.head() #shows the first 5 rows
```

```python
energy_df["Country"].value_counts() #counts how many values there are for each country
```

```python
energy_df.duplicated().sum() #sums the duplicated rows 
```

```python
usa_renewable_df = energy_df[energy_df["Country"] == "USA"] #gets the rows from country column that are USA
avg_renewable_usa = usa_renewable_df["Renewable Energy (%)"].mean() #gets the average of the renewable energy column
print(avg_renewable_usa) #prints the average
```

```python
energy_df.loc[energy_df["Country"] == "USA", "Renewable Energy (%)"] = energy_df.loc[energy_df["Country"] == "USA", "Renewable Energy (%)"].fillna(22.927072487674756)
#fills the NaN values with the average value we got
```

```python
energy_df.loc[energy_df["Country"] == "USA", "Renewable Energy (%)"] #locates the renewable energy column
```

```python
energy_df #shows the whole dataframe
```

```python
energy_df.isnull().sum() #another check for null values
```

```python
countries = ["USA", "China", "India", "Germany", "Brazil", "Australia", "South Africa", "Canada", "Japan", "UK"] #list of countries in dataframe
for i in countries: #iterates through countries
    renewable_df = energy_df[energy_df["Country"] == i] #gets the countries in dataframe 
    avg_renewable = renewable_df["Renewable Energy (%)"].mean() #gets the average for each countrys corresponding renewable energy values
    energy_df.loc[energy_df["Country"] == i, "Renewable Energy (%)"] = energy_df.loc[energy_df["Country"] == i, "Renewable Energy (%)"].fillna(avg_renewable)
    #fills all NaN values in renewable energy column with the specific averages for countries
```

```python
energy_df.isnull().sum() #another check for null values
```

```python
countries = ["USA", "China", "India", "Germany", "Brazil", "Australia", "South Africa", "Canada", "Japan", "UK"]
columns = ["Government Investment (Million USD)", "Solar", "Wind", "Hydro", "Geothermal", "Biomass", "Other Renewables"] #list of countries and columns
for i in countries: #iterates through countries
    for col in columns: #iterates through columns
        nanvalues_df = energy_df[energy_df["Country"] == i] #gets the country name
        avg_of_column = nanvalues_df[col].mean() #gets the average of the column that is iterated through
        energy_df.loc[energy_df["Country"] == i, col] = energy_df.loc[energy_df["Country"] == i, col].fillna(avg_of_column) 
        #fills the NaN values in the iterated column for its specific average and specific country
```

```python
energy_df.isnull().sum() #last check for null values
```

```python
energy_df.to_csv("Renewable_Energy_Adoption.csv", index=False) #writes changes to the csv file
```

```python
usa_data = energy_df[energy_df["Country"] == "USA"] #gets the total energy consumption of usa

plt.figure(figsize=(10, 6)) #creates a figure of size 10 to 6
#plots a line graph with years as x values and energy consumption as y values for usa
plt.plot(usa_data["Year"], usa_data["Total Energy Consumption (TWh)"], marker="o", color="blue", label="USA") 
plt.title("Total Energy Consumption - USA") #gives the figure a title
plt.xlabel("Year") #x axis label 
plt.ylabel("Energy Consumption (TWh)") #y axis label
plt.legend() #legend for the line
plt.show() #shows the plot
```

```python
usa_data = energy_df[energy_df["Country"] == "USA"] #gets data of USA, UK and China
china_data = energy_df[energy_df["Country"] == "China"]
uk_data = energy_df[energy_df["Country"] == "UK"]

plt.figure(figsize=(10, 8)) #creates a figure of size 10 to 8 to plot the increase of renewable energy use of USA, UK and China
plt.plot(usa_data["Year"], usa_data["Renewable Energy (%)"], marker="o", color="blue", label="USA") #plots line for usa
plt.plot(china_data["Year"], china_data["Renewable Energy (%)"], marker="o", color="orange", label="China") #plots line for china
plt.plot(uk_data["Year"], uk_data["Renewable Energy (%)"], marker="o", color="red", label="UK") #plots line for UK
plt.title("Use of Renewable Energy - USA") #title of graph
plt.xlabel("Year") #x axis label
plt.ylabel("Renewable Energy (%)") #y axis label
plt.legend() #legend to seperate countries with line colours
plt.show() #shows the graph
```

```python
energy_df["Government Investment (Million USD)"] = energy_df["Government Investment (Million USD)"].astype(int).astype(str) + ".0" 
energy_df["Government Investment (Million USD)"] = energy_df["Government Investment (Million USD)"].astype(float)
#fixes the government investment column because loop gave decimal values for millions of USD values
```

```python
countries = ["USA", "China", "India", "Germany", "Brazil", "Australia", "South Africa", "Canada", "Japan", "UK"] #list of countries
sumlist = [] #empty list to store each countrys sum over the years in order
for i in countries: #iterates over the countries
    investment_df = energy_df[energy_df["Country"] == i] #gets the country 
    sum_investment = investment_df["Government Investment (Million USD)"].sum() #sums the values of investments for the country
    sumlist.append(sum_investment) #adds the sum to the list
plt.figure(figsize=(10,6))
plt.bar(countries, sumlist) #plots a bar chart
plt.show() #shows it
sumlistreduction = [] #empty list for emission reduction sums
for i in countries: #iteraets over countries
    investment_df = energy_df[energy_df["Country"] == i] #gets country
    sum_investment = investment_df["Emissions Reduction (%)"].sum() #gets the sum in emission reduction of the country
    sumlistreduction.append(sum_investment) #adds the sum
plt.figure(figsize=(10,6))
plt.bar(countries, sumlistreduction) #plots a bar chart
plt.show() #shows it
```

```python
#list of columns that are types of renewable energies and the Emissions Reduction rates
columns_to_correlate = ["Emissions Reduction (%)", "Solar", "Wind", "Hydro", "Geothermal", "Biomass", "Other Renewables"]
selected_df = energy_df[columns_to_correlate] #gets the column names

corr = selected_df.corr() #correlates between columns to see if they correlate

plt.figure(figsize=(10, 8)) #creates a figure of size 10 to 8 for a heatmap that shows correlations of columns
sb.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5) #creates a heatmap 
plt.title('Correlation Heatmap between Renewable Energy Types and Emissions Reduction Rates', fontsize=10) #creates a title for heatmap
plt.show() #shows the heatmap
```
