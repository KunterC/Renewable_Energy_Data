```python
usa_renewable_df = energy_df[energy_df["Country"] == "USA"]
avg_renewable_usa = usa_renewable_df["Renewable Energy (%)"].mean()
print(avg_renewable_usa)
```

```python
energy_df.loc[energy_df["Country"] == "USA", "Renewable Energy (%)"] = energy_df.loc[energy_df["Country"] == "USA", "Renewable Energy (%)"].fillna(22.927072487674756)
```

```python
countries = ["USA", "China", "India", "Germany", "Brazil", "Australia", "South Africa", "Canada", "Japan", "UK"]
for i in countries:
    renewable_df = energy_df[energy_df["Country"] == i]
    avg_renewable = renewable_df["Renewable Energy (%)"].mean()
    energy_df.loc[energy_df["Country"] == i, "Renewable Energy (%)"] = energy_df.loc[energy_df["Country"] == i, "Renewable Energy (%)"].fillna(avg_renewable)
```

```python
energy_df.isnull().sum()
```

```python
countries = ["USA", "China", "India", "Germany", "Brazil", "Australia", "South Africa", "Canada", "Japan", "UK"]
columns = ["Government Investment (Million USD)", "Solar", "Wind", "Hydro", "Geothermal", "Biomass", "Other Renewables"] 
for i in countries:
    for col in columns:
        renewable_df = energy_df[energy_df["Country"] == i]
        avg_renewable = renewable_df[col].mean()
        energy_df.loc[energy_df["Country"] == i, col] = energy_df.loc[energy_df["Country"] == i, col].fillna(avg_renewable)
```

```python
energy_df.isnull().sum()
```

```python
energy_df.to_csv("Renewable_Energy_Adoption.csv", index=False)
```

```python
usa_data = energy_df[energy_df["Country"] == "USA"]

plt.figure(figsize=(10, 6))
plt.plot(usa_data["Year"], usa_data["Total Energy Consumption (TWh)"], marker="o", color="blue", label="USA")
plt.title("Total Energy Consumption - USA")
plt.xlabel("Year")
plt.ylabel("Energy Consumption (TWh)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
usa_data = energy_df[energy_df["Country"] == "USA"]
china_data = energy_df[energy_df["Country"] == "China"]
uk_data = energy_df[energy_df["Country"] == "UK"]

plt.figure(figsize=(10, 6))
plt.plot(usa_data["Year"], usa_data["Renewable Energy (%)"], marker="o", color="blue", label="USA")
plt.plot(china_data["Year"], china_data["Renewable Energy (%)"], marker="o", color="orange", label="China")
plt.plot(uk_data["Year"], uk_data["Renewable Energy (%)"], marker="o", color="red", label="UK")
plt.title("Use of Renewable Energy - USA")
plt.xlabel("Year")
plt.ylabel("Renewable Energy (%)")
plt.legend()
plt.grid(True)
plt.show()
```
