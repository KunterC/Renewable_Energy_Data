```python
countries = ["USA", "China", "India", "Germany", "Brazil", "Australia", "South Africa", "Canada", "Japan", "UK"]
for i in countries:
    renewable_df = energy_df[energy_df["Country"] == i]
    avg_renewable = renewable_df["Renewable Energy (%)"].mean()
    energy_df.loc[energy_df["Country"] == i, "Renewable Energy (%)"] = energy_df.loc[energy_df["Country"] == i, "Renewable Energy (%)"].fillna(avg_renewable)
```

```python
energy_df.isnull().sum()
```

```python
countries = ["USA", "China", "India", "Germany", "Brazil", "Australia", "South Africa", "Canada", "Japan", "UK"]
columns = ["Government Investment (Million USD)", "Solar", "Wind", "Hydro", "Geothermal", "Biomass", "Other Renewables"] 
for i in countries:
    for col in columns:
        renewable_df = energy_df[energy_df["Country"] == i]
        avg_renewable = renewable_df[col].mean()
        energy_df.loc[energy_df["Country"] == i, col] = energy_df.loc[energy_df["Country"] == i, col].fillna(avg_renewable)
```
